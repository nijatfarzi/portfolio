# EXTRACT

```python
import pandas as pd
df = pd.read_excel("∆èsas data.xlsx")
df.head(5)

df["Order Date"] = pd.to_datetime(df["Order Date"])
```

# RFM

```python
current_day = df["Order Date"].max() + pd.Timedelta(days=1)
```

```python
rfm_df = df.groupby("Customer ID").agg({
    "Order Date" : lambda x: (current_day - x.max()).days,
    "Order ID" : "nunique",
    "Sales" : "sum"
}).reset_index()

rfm_df.columns = ["Customer ID","Recency","Frequency","Monetary"]
```

# ROI

```python
rfm_df["Cost"] = 2000
rfm_df["ROI"] = ((rfm_df["Monetary"] - rfm_df["Cost"]) / rfm_df["Cost"]).round()
rfm_df["ROI_Percent"] = rfm_df["ROI"] * 100
rfm_df.head()
```

# Churn Rate

```python
churn_limit = 90
total_customers = rfm_df.shape[0]
churned_customers = rfm_df[rfm_df["Recency"] > churn_limit].shape[0]
rfm_df["Churn_Rate"] = round((churned_customers / total_customers) * 100)
rfm_df.head()
```
