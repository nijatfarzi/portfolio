````python
import pandas as pd
import numpy as np
df = pd.read_csv("HR Data.csv")
````



## 🔥 Yeni Mini Görevler (HR Data.csv ile)

### 🔹 Görev 1: **Hangi eğitim alanında (EducationField) çalışanların maaş ortalaması en yüksektir?**
```python
# 1. Her EducationField için maaş ortalamasını hesapla
```

### 🔹 Görev 2: **Yaşı 40'tan büyük olan çalışanlar arasında en uzak mesafeden gelen ilk 5 kişiyi göster.**

### 🔹 Görev 3: **Evli çalışanların (MaritalStatus = "Married") ortalama iş memnuniyeti nedir?**
