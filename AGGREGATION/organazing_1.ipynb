```python
import pandas as pd
df = pd.read_excel("factory_data.xlsx")

#Soru: Revenue sütununa göre azalan sıralama yap, ilk 10 satırı göster.
df["Revenue"].sort_values(ascending= False).head(10) #Sadece Revenue sutunun gosterir
df.sort_values(by="Revenue", ascending= False).head(10) 

# Soru: Index’i artan ve azalan şekilde sırala, farkını gör.
df.sort_index()
df.sort_index(ascending= False) #tersden

# Soru: Index’i sıfırla ama önce drop=True ile eski index’i yok et, farkını incele.

df.reset_index(drop=True)
```

```python

# Soru: “Month” sütununu index yap, sonra index sırasına göre sırala.
df.set_index("Month").sort_index(ascending=False)

# Soru: “Revenue” sütununu “Gelir” olarak değiştir.

df.rename(columns={"Revenue" : "Gelir"})

# Soru: Tüm sütun adlarını küçük harfe çevir.
import string as st
df.columns = list(st.ascii_lowercase[:len(df.columns)])
df.columns
```
