```python
import pandas as pd
df = pd.read_excel("factory_data.xlsx")
df.head()

# âœ… 1. sum() â†’ Toplama
# ðŸ‘‰ df["Revenue"].sum() â†’ Revenue toplamÄ±
# ðŸ‘‰ df.groupby("Category")["Revenue"].sum() â†’ Kategoriya bazda Revenue toplamÄ±
df.groupby("Month")["Sales"].sum()
month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
               "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

df["Month"] = pd.Categorical(df["Month"], categories=month_order, ordered=True)
df = df.sort_values("Month")
df.head()
```

```python

# âœ… 2. mean() â†’ Ortalama
# ðŸ‘‰ df["Revenue"].mean() â†’ Ortalama Revenue
# ðŸ‘‰ df.groupby("Category")["Revenue"].mean() â†’ Kategoriya bazda ortalama
df.groupby("Category")["Sales"].mean().round()
```

```python

# âœ… 3. count() â†’ Sayma
# ðŸ‘‰ df["Category"].count() â†’ Categoryâ€™de boÅŸ olmayan setÄ±r sayÄ±sÄ±
# ðŸ‘‰ df.groupby("Category")["Revenue"].count() â†’ Kategoriy bazda Revenue sayÄ±sÄ±

df.groupby("Month")["Revenue"].count().reset_index(name="Revenue_count")
```

```python
# âœ… 4. min(), max() â†’ Minimum ve maksimum
# ðŸ‘‰ df["Revenue"].min(), df["Revenue"].max() â†’ En kiÃ§ik, en boyÃ¼k Revenue

# âœ… 5. describe() â†’ statistik Ã¶zeti
# ðŸ‘‰ df.describe() â†’ Butun numeric sÃ¼tunlarÄ±n count, mean, std, min, 25%, 50%, 75%, max Ã¶zetini verir

# âœ… 6. agg() â†’ Ã–zel Ã¶zetleme
# ðŸ‘‰ df.agg({"Revenue": ["sum", "mean"], "Expenses": ["max"]})
df.agg({"Month" : "count", "Sales" :["mean", "std"]}).T.fillna(0).round()
```

```python

# âœ… 7. groupby() + agg() â†’ Kombo Ã¶zet
# ðŸ‘‰ df.groupby("Category").agg({"Revenue": "sum", "Expenses": "mean"})

df.groupby("Month").agg({"Revenue" : "mean", "Sales" : ["std","sum"]}).round()


# âœ… 8. pivot_table() â†’ Pivot table Ã¶zetleme
# ðŸ‘‰ df.pivot_table(values="Revenue", index="Category", aggfunc="sum")

df.pivot_table(index ="Month", columns="Category", values="Revenue", aggfunc=["sum","mean"])
```
