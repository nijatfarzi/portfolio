```python
import pandas as pd

df = pd.read_csv("HR Data.csv")
df.head(5)
```

```python
df.groupby(["Department", "Gender"]).agg(
    Mean = ("MonthlyIncome","mean"),
    Max = ("JobSatisfaction","max")
).round(1)
```

```python
pd.pivot_table(df,index="Gender", columns="JobRole",values="MonthlyIncome", aggfunc="mean").round(2)
```

```python

performance = pd.DataFrame({
    "EmployeeNumber": [1, 2, 5],
    "Rating": [4, 3, 2]
})
df_new = pd.merge(df, performance, on = "EmployeeNumber", how="left")
df_new[df_new["Rating"] == 4]
```

```python

result =pd.pivot_table(df, index="JobRole", columns="Gender", 
                       aggfunc={"MonthlyIncome":"mean","Age":"median"}
                       ).rename(columns={"MonthlyIncome":"AvgIncome","Age" : "MedianAge"}
                                ).round(2)
result
    
```

```python
qrup = df.groupby("Department")["MonthlyIncome"].mean().reset_index() # reset olmasa olmaz
qrup[qrup["MonthlyIncome"] > 6000].round(2) 
```
