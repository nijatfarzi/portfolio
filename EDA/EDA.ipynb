```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("HR Data.csv")
df.head(5)

numeric_df = df.select_dtypes(include="number")
corr = numeric_df.corr()

plt.figure(figsize=(8,6))
sn.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Colerasiyon map")
plt.show()

# Scatterplot 
plt.figure(figsize=(8,6))
sn.scatterplot(data=df, x= "MonthlyIncome", y="TotalWorkingYears", hue="Attrition")
plt.title("Total Working Years vs Monthly Income")
plt.xlabel("Total Working Years")
plt.ylabel("Monthly Income")
plt.show()

# Pairplot 
sn.pairplot(numeric_df)
plt.show()

plt.figure(figsize=(8,6))
sn.scatterplot(data=df, x="Age", y="TotalWorkingYears")

# Doruyu iz: Y = X
max_age = df["Age"].max()
x = np.linspace(0, max_age, 100)
plt.plot(x, x, color='red', linestyle='--')

plt.title("Age vs Total Working Years")
plt.xlabel("Age")
plt.ylabel("Total Working Years")
plt.show()
```
