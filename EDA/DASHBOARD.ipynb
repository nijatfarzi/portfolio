```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [100, 120, 150, 130, 160, 180],
    'Region': ['North', 'South', 'East', 'West'],
    'Revenue': [400, 600, 550, 450],
    'Category': ['A', 'B', 'C'],
    'CategoryShare': [40, 35, 25],
    'MarketingSpend': [10, 20, 30, 40, 50, 60],
    'RevenueGrowth': [100, 150, 200, 220, 280, 310]
}

# # LINE CHART: Sales trend
plt.figure(figsize=(6,4))
plt.plot(data['Month'], data['Sales'], marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

# # PIE CHART: Category share
plt.figure(figsize=(4,4))
plt.pie(data['CategoryShare'], labels=data['Category'], autopct='%1.1f%%')
plt.title("Revenue Share by Category")
plt.show()

# # BAR CHART: Region-wise revenue
plt.figure(figsize=(6,4))
plt.bar(data['Region'], data['Revenue'], color='skyblue')
plt.title("Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Revenue")
plt.show()

# # SCATTER PLOT: Marketing Spend vs Revenue Growth
plt.figure(figsize=(6,4))
sns.scatterplot(x=data['MarketingSpend'], y=data['RevenueGrowth'])
plt.title("Marketing Spend vs Revenue Growth")
plt.xlabel("Marketing Spend")
plt.ylabel("Revenue Growth")
plt.grid(True)
plt.show()
```

```python
# LINE CHART: Sales trend
plt.figure(figsize=(6,4))
plt.plot(data["Month"], data["Sales"], marker="o")
plt.title("Sales over Month")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

```
