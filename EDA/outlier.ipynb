import pandas as pd

seri = pd.Series([1, 2, 3, 4, 5, 6, 50])

````python

# ## ğŸ¯ Outlier AlÄ±ÅŸtÄ±rmasÄ±: â€œÅirkette UÃ§anlarâ€ ğŸš€

# **Konu:** HR Data.csv  
# **Hedef:** Åirkette diÄŸer Ã§alÄ±ÅŸanlardan **Ã§ok farklÄ± davranan (outlier)** Ã§alÄ±ÅŸanlarÄ± bulmak

# ---
import pandas as pd

df = pd.read_csv("HR Data.csv")
# ### ğŸ§ª GÃ¶revler:

# #### ğŸ”¹ GÃ¶rev 1:
# `TotalWorkingYears` sÃ¼tunundaki aykÄ±rÄ± (outlier) deÄŸerleri **IQR yÃ¶ntemiyle** tespit et.

# - Q1, Q3 ve IQR deÄŸerlerini hesapla  
# - Alt ve Ã¼st sÄ±nÄ±rlarÄ± bul  
# - Bu sÄ±nÄ±rlara gÃ¶re outlier olan satÄ±rlarÄ± filtrele ve yazdÄ±r

q1 = df["TotalWorkingYears"].quantile(0.25)
q3 = df["TotalWorkingYears"].quantile(0.75)
iqr = q3 -q1

alt = q1 - 1.5 * iqr
ust = q3 + 1.5 * iqr

outlier = df[(df["TotalWorkingYears"] < alt ) | (df["TotalWorkingYears"] >  ust )]
outlier

# #### ğŸ”¹ GÃ¶rev 2:
# BulduÄŸun outlierâ€™larÄ±n sadece ÅŸu bilgilerini yazdÄ±r:
# - `Age`, `TotalWorkingYears`, `MonthlyIncome`
outlier[["Age", "TotalWorkingYears", "MonthlyIncome"]]

# #### ğŸ”¹ GÃ¶rev 3:
# Outlier Ã§alÄ±ÅŸanlarÄ±n ortalama maaÅŸÄ± ile normal Ã§alÄ±ÅŸanlarÄ±n ortalama maaÅŸÄ±nÄ± karÅŸÄ±laÅŸtÄ±r.
outliermask = (df["TotalWorkingYears"] < alt ) | (df["TotalWorkingYears"] >  ust )
outlier_olmayanlar = df[~outliermask]
outlier_olmayanlar["MonthlyIncome"].mean() 
outlier["MonthlyIncome"].mean()

# ```python
# # Ã¶rnek fikir:
# outlier_olmayanlar = df[~(outlier filtresi)]
# ```

# #### ğŸ”¹ GÃ¶rev 4 (Bonus):
# Outlier Ã§alÄ±ÅŸanlar genelde genÃ§ mi yaÅŸlÄ± mÄ±? OrtalamalarÄ±nÄ± karÅŸÄ±laÅŸtÄ±r ve yorum yap.
# ğŸ¯ GÃ¶rev AdÄ±mlarÄ±:
# 1. Daha Ã¶nce oluÅŸturduÄŸun outlier ve outlier_mask deÄŸiÅŸkenlerini kullan.
# 2. Åu iki ortalamayÄ± karÅŸÄ±laÅŸtÄ±r:
outlier = outlier["Age"].mean().round(2)
inlier = outlier_olmayanlar["Age"].mean().round(2)

outlier > inlier

# 3. Sonucu yazdÄ±r ve yorumla:
print("Outlier: " , outlier)
print("Outlier olmayan: ", inlier)
# Hangi grup daha yaÅŸlÄ±?# Sence neden?
# outlier ortalamasi dah fazla cunku normaldan fazla caliwmiwsa daha onceden iwe alinmiw, yani daha yasli

# ---

# ## ğŸ“Œ Ä°pucu:
# AynÄ± mantÄ±ÄŸÄ± kullan:
# ```python
# q1 = df["TotalWorkingYears"].quantile(0.25)
# q3 = df["TotalWorkingYears"].quantile(0.75)
# iqr = q3 - q1
# alt_sinir = q1 - 1.5 * iqr
# ust_sinir = q3 + 1.5 * iqr
# ```

````
