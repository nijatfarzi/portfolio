```python

 ## 🔹 A. **Correlation Matrix (Korelasyon Matrisi)**

### 📘 Teori:
# - Sayısal sütunlar arasındaki doğrusal ilişkiyi ölçer.  
# - Korelasyon katsayısı `-1 ile +1` arasında değer alır:  
#   - `+1` → Güçlü pozitif ilişki  
#   - `0` → Hiçbir ilişki yok  
#   - `-1` → Güçlü negatif ilişki  
# - Negatif ilişki: biri artarken diğeri azalır.  
# - Pozitif ilişki: ikisi birlikte artar veya azalır. -->


### 💻 Python Örneği:

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Örnek veri seti (Hazır varsa onu da kullanabiliriz)
df = pd.read_csv("HR Data.csv")
df.head(5)

#  Korelasyon matrisi
numeric_df = df.select_dtypes(include='number')
corr_matrix = numeric_df.corr()

# Isı haritası ile görselleştir
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()

```
