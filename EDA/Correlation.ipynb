```python

 ##  A. **Correlation Matrix (Korelasyon Matrisi)**

###  Teori:
# - Saysal stunlar arasndaki dorusal ilikiyi ler.  
# - Korelasyon katsays `-1 ile +1` arasnda deer alr:  
#   - `+1`  Gl pozitif iliki  
#   - `0`  Hibir iliki yok  
#   - `-1`  Gl negatif iliki  
# - Negatif iliki: biri artarken dieri azalr.  
# - Pozitif iliki: ikisi birlikte artar veya azalr. -->


###  Python rnei:

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# rnek veri seti (Hazr varsa onu da kullanabiliriz)
df = pd.read_csv("HR Data.csv")
df.head(5)

#  Korelasyon matrisi
numeric_df = df.select_dtypes(include='number')
corr_matrix = numeric_df.corr()

# Is haritas ile grselletir
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()

```
